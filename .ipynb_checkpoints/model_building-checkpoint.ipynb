```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')

%matplotlib inline
```

```python
data=pd.read_csv("co2.csv")
```

```python
data.head()
```

```python
data.info()
```

we imported the csv file and notice that the file contains the yyyymm so we need to separate the year and month

```python
data['year']=data.YYYYMM.astype(str).str[0:4].astype(float)
data['month']=data.YYYYMM.astype(str).str[4:7].astype(float)


data.drop('YYYYMM',axis=1,inplace=True)
data.replace([np.inf, -np.inf], np.nan, inplace=True)
```

```python
data.tail()
```

# Missing values

```python
data.isnull().sum()
```

Hence we obsevere that there are not missing values

```python
data.info()
```

```python
for feature in data.columns:
    sns.distplot(x=data[feature])
    plt.xlabel(feature)
    plt.show()
```

```python
plt.figure(figsize=(6,4),dpi=120)
sns.lineplot(x='year',y='Value',data=data)
```

```python
data.corr()
sns.heatmap(data.corr(),annot=True)
plt.show()
```

Here from the heatmap we observe that the year and value are correlated

```python
data.columns
```

```python
X=data.drop('Value',axis=1)
y=data['Value']
```

```python
X.head()
```

```python
import xgboost as xgb
data_dmatrix=xgb.DMatrix(X,label=y)
data_dmatrix
```

# Train and test split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.12,random_state=0)
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

# Model building

```python
import xgboost as xgb
```

```python
reg_mod = xgb.XGBRegressor(
    n_estimators=1000,
    learning_rate=0.08,
    subsample=0.75,
    colsample_bytree=1, 
    max_depth=7,
    gamma=0,
)
```

```python
reg_mod.fit(X_train,y_train)
```

```python
#after training the model we check the scores of the model
from sklearn.model_selection import cross_val_score, KFold
scores=cross_val_score(reg_mod,X_train,y_train,cv=10)
```

```python
scores.mean()
```

```python
prediction=reg_mod.predict(X_test)
```

# Evaluation

```python
from sklearn.metrics import r2_score
r2 = np.sqrt(r2_score(y_test, prediction))
print("R_Squared Score : %f" % (r2))
```

As you can see, the these statistical metrics have reinstated our confidence about this model. RMSE ~ 4.95 R-Squared Score ~ 98.8% Now, let's visualize the original data set using the seaborn library.

### let visualize the test and predicted data

```python
plt.figure(figsize=(8,6),dpi=80)
ax=range(len(y_test))
plt.plot(ax,y_test,color='blue',label="test")
plt.plot(ax,prediction,color='red',label="predicted")
plt.title('test and predicted data')
plt.legend()
plt.show()
```

```python
df=pd.DataFrame(prediction, columns=['pred']) 
df['date'] = pd.date_range(start='8/1/2016', periods=len(df), freq='M')
```
